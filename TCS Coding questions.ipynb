#####  Write a C program to calculate the factorial of a non-negative integer N. The factorial of a number N is defined as the product of all integers from 1 up to N. Factorial of 0 is defined to be 1. The number N is a non-negative integer that will be passed to the program as the first command line parameter. Write the output to stdout formatted as an integer WITHOUT any other additional text. You may assume that the input integer will be such that the output will not exceed the largest possible integer that can be stored in an int type variable

```python
def fact(n):
    if n == 1 or n==0:
        return 1
    if n >= 2:
        return n*fact(n-1)
n = int(input())
print(fact(n))
```

#### Given an input string, you have to encrypt the string in such a manner that a becomes z, z becomes a and accordingly.
#### Note : All the Characters in the given string are in lowercase alphabets

```python
s = input()
l = []
str1 = ""
for i in s:
    i = ord('a')-ord(i)+ord('z')
    l.append(chr(i))
for j in l:
    str1+=j
print(str1)
```

#### Oddly Even Problem Statement Given a maximum of 100 digit numbers as input, find the difference between the sum of odd and even position digits

```python
num = [int(d) for d in str(input("Enter the number"))]
odd,even = 0,0
for i in range(0,len(num)):
    if i%2 == 0:
        even += num[i]
    else:
        odd += num[i]
print(abs(odd-even))
```

#### A cricket coach wants to measure the area of cricket ground and he knows the Radius of the cricket ground 
#### Given Formula : P= pi*r^2, where pi = 3.1415
#### Constraints:  r<30 and r>20
#### If he enters any number less than 20 and greater than 30 it should display "Wrong Display Output")
#### Input : If he enters any number between 20 and 30 , the program should calculate the area and prints the area as output

```python
def area(r):
    ar = 3.1415*(r**2)
    return ar
r = float(input())
if r<20 and r>30:
    print("Wrong Display Output")
else:
    print("%.3f"%area(r))
```

```python
def isArmstrong(n):
    sum1 = 0
    temp = n
    l = len(str(n))
    while temp!=0:
        digit = temp%10
        sum1+=digit ** l
        temp = temp//10
    if sum1 == n:
        return True
    else:
        return False
N1 = int(input())
N2 = int(input())
if N1<=0 or N2<=0:
    print("Wrong Input")
else:
    for i in range(N1,N2+1):
        if isArmstrong(i):
            print(i)

            
```

####  Sum of First N natural numbers

```python
def sumOfNaturalNumbers(n):
    sum1 = 0
    x =1
    while x<=n:
        sum1 = sum1+x
        x+=1
    return sum1
#driver code
n = int(input())
print(sumOfNaturalNumbers(n))
```

## Leap year or not

```python
y = int(input())
if y%4==0:
    if y%100==0:
        if y%400==0:
            print("Leap Year")
        else:
            print("Not a leap Year")
    else:
        print("Leap Year")
else:
    print("Not a leap Year")
```

## Prime Numbers

```python
num = int(input())
if num>1:
    for i in range(2,int(num/2)+1):
        if num%i ==0:
            print(num,"not a prime number")
            break
    else:
        print(num,"is a prime number")
else:
    print(num,"not a prime number")
```

## Sum of digits of a number

```python
def sumOfDigits(n):
    sum1 = 0
    for i in str(n):
        sum1 += int(i)
    return sum1
n = input()
print(sumOfDigits(n))
```

## Reverse of a number

```python
n = int(input())
rev = 0
while(n>0):
    rem = n%10
    rev = rev*10+rem
    n=n//10
print(rev)
```
